```python
import numpy as np
import pickle
import matplotlib.pyplot as plt
```

```python
berry_3D_nosym = pickle.load(open("Te_berry.pickle","rb"))

vo =20
vdo = 10

NK = 60
dx, dy = 1./NK, 1./NK
x, y = np.mgrid[slice(-0.5, 0.5 + dy, dy),
                slice(-0.5, 0.5 + dx, dx)]

res_berry=np.zeros((NK,NK,1))
res_der=np.zeros((NK,NK,1))
res_E = np.zeros((NK,NK,1))

Ef=5.85
for ib in range(0,20):
    band=berry_3D_nosym.get_data(iband=ib,quantity='E',component=None)-Ef
    berry=berry_3D_nosym.get_data(iband=ib,quantity='berry',component=None)
   # der=berry_3D_nosym.get_data(iband=ib,quantity='Der_berry',component=None)
    select = band < -0.25
    res_E[select] += band[select]
    res_berry[select] += berry[:,:,:,2][select]
   # res_der[select] += der[:,:,:,2,2][select]
    
c = plt.pcolormesh(x, y, res_berry[:,:,0], cmap ='seismic',vmax=vdo,vmin=-vdo)
plt.colorbar(c)

fig = plt.figure(figsize=(14,10))
```
